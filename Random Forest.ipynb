```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from model_building import naive_model, tuned_model, evaluate_model

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the .npz file
data = np.load('data.npz')
```

```python
# Access the arrays
X_train_balanced = data['X_train_balanced']
y_train_balanced = data['y_train_balanced']
X_valid_scaled = data['X_valid_scaled']
y_valid = data['y_valid']
```

```python
rf_model = naive_model(RandomForestClassifier(), X_train_balanced, y_train_balanced)
evaluate_model(rf_model, X_valid_scaled, y_valid)
```

```python
rf_param_grid = {'n_estimators': [100, 200, 300], 'max_depth': [10, 20, 30], 'min_samples_split': [2, 5, 10]}
rf_grid = tuned_model(RandomForestClassifier(), rf_param_grid, X_train_balanced, y_train_balanced)
evaluate_model(rf_grid, X_valid_scaled, y_valid)
```
