```python
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from model_building import naive_model, tuned_model, evaluate_model

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the .npz file
data = np.load('data.npz')
```

```python
# Access the arrays
X_train_balanced = data['X_train_balanced']
y_train_balanced = data['y_train_balanced']
X_valid_scaled = data['X_valid_scaled']
y_valid = data['y_valid']
```

```python
adaboost_model = naive_model(AdaBoostClassifier(), X_train_balanced, y_train_balanced)
evaluate_model(adaboost_model, X_valid_scaled, y_valid)
```

```python
adaboost_param_grid = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 1.0]}
adaboost_grid = tuned_model(AdaBoostClassifier(), adaboost_param_grid, X_train_balanced, y_train_balanced)
evaluate_model(adaboost_grid, X_valid_scaled, y_valid)
```
